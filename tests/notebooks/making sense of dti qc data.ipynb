```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
swu_data = pd.read_csv('swu_quality_vector.csv')
groups = swu_data.groupby('quality')
```

```python
# Plot
fig, ax = plt.subplots()
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for name, group in groups:
    ax.plot(group.mse, group.hellinger, marker='o', linestyle='', ms=12, label=name)
fig.set_size_inches(10.5, 8.5)
ax.legend(('bad', 'good'))
plt.xlabel('mse')
plt.ylabel('hellinger distance')
```

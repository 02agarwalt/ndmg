```python
def normalize_bvecs(bvecs):
    for i, k in zip(bvecs, range(len(bvecs))):
        norm = np.linalg.norm(i)
        if norm != 1.0 and norm != 0.0:
            bvecs[k, :] = np.divide(bvecs[k, :], norm)
```
